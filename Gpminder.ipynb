To get a first look at the state of global agriculture we call the gapminder method defined previously for the most recent year available in our given dataset, 2019.

The Gapminder above charts the relationship between the quantity of fertilizer a country uses, plotted on the x-axis as well as being further empasized through a darkening color gradient, and the total output produced, plotted on the y-axis. As a third variable we have chosen land quantity which is represented by the area of each point displayed on the chart. First glance shows that there is a strong positive relationship between the amount of fertilizer a country uses and the output they produce, which is expected as fertilisers are additaves designed to increase the productivity of crop yields. The Gapminder also shows a slight trend that an increase in the amount of land available has a postive effect on outputs as the area of points generally increases as output quantity increases. More land implies more space to farm. This rule is not as strong though as we can see there are several smaller countries generating greater outputs through the use of more fertalizers.
